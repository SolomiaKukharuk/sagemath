## Завдання 1

Розгляньте групу матриць за множенням  

$$G = \left\langle \left(\begin{array}{rrr}
1 & 0 & 1 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{array}\right),  \left(\begin{array}{rrr}
1 & 0 & 0 \\
0 & 1 & 1 \\
0 & 0 & 1
\end{array}\right)\right\rangle$$

Дайте відповідь на наступні питання: 

1. _Чи є група G скінченною?_
2. _Чи є група G комутативною?_
3. _Який вигляд мають елементи групи G?_
4. _Які порядки мають елементи групи G?_ 
5. _Чому ізоморфна група G?_


__Відповіді__: 
1. Ні
2. Так
3. Елементи будуть мати вигляд {[1,0,0],[0,1,n],[0,0,1], де n це додатнє натуральне число}
4. Нескінченність
5. ізоморфна прямому добутку циклічних груп

```python
from sage.all import *
A = Matrix([[1,0,1],[0,1,0],[0,0,1]])
B = Matrix([[1,0,0],[0,1,1],[0,0,1]])
G = MatrixGroup([A,B])
#1
print('G is finite:',G.is_finite())
#2
print("G is abelian:",G.is_abelian())
#3,4
#Так,як група нескінченна можемо взяти лише деякі її елементи
#Всі елементи групи це будуть добутки степенів генераторів, або самі їхні степені
elements = print([A**i * B**j for i in range(5) for j in range(5)])  # Вибираємо обмежену кількість
def finite_order(matrix, max_iter=1000):
    identity = matrix.parent().identity_matrix()
    power = matrix
    order = 1
    while power != identity:
        power *= matrix
        order += 1
        if order > max_iter:  # обмеження для нескінченних порядків
            return 'inf'
    return order
element_orders = [(g, finite_order(g)) for g in elements]

print(element_orders)
#5
#Якщо група комутативна, то вона буде ізоморфна прямому добутку циклічних груп (інколи нескінченних циклічних груп).
#Комутативність можна перевірити на генераторах
# Перевірка комутативності
is_commutative = (A*B == B*A)
print('G is commutative:',is_commutative)
```

```python
from sage.all import *
A = Matrix([[1,-1],[-1,2]])
B = Matrix([[3,2],[2,1]])
G = MatrixGroup([A,B])
#1
print('G is finite:',G.is_finite())
#2
print("G is abelian:",G.is_abelian())
#5
is_commutative = (A*B == B*A)
print('G is commutative:',is_commutative)
#3,4
elements = print([A**i * B**j for i in range(5) for j in range(5)])  # Вибираємо обмежену кількість
def finite_order(matrix, max_iter=1000):
    identity = matrix.parent().identity_matrix()
    power = matrix
    order = 1
    while power != identity:
        power *= matrix
        order += 1
        if order > max_iter:  # обмеження для нескінченних порядків
            return 'inf' 
    return order
element_orders = [(g, finite_order(g)) for g in elements]

print(element_orders)
```

__Додаткова задача:__
Опишіть підгрупи групи K з точністю до ізоморфізму. 

```python
from sage.all import *

# Визначаємо матриці A і B
A = Matrix([[-1, 0, 0], [0, 1, 1/2], [0, 0, 1]])
B = Matrix([[1, 0, 1], [0, 1, 0], [0, 0, 1]])

# Створюємо групу G
G = MatrixGroup([A, B])

# 1
elements = []
for i in range(3): 
    for j in range(3):
        elements.append(A**i * B**j)

unique_elements = set(tuple(g.list()) for g in elements)


is_finite = len(unique_elements) < float('inf')
print('G is finite:', is_finite)

# 2
is_commutative = (A * B == B * A)
print('G is commutative:', is_commutative)

# 3
if is_finite:
    for g in unique_elements:
        element_matrix = Matrix(g)
        print(element_matrix)  # Перетворюємо назад в матрицю для виводу

# 4
def compute_order(matrix, max_iterations=100):
    """Обчислює порядок матриці."""
    # Перевіряємо, чи матриця квадратна
    if matrix.nrows() != matrix.ncols():
        return float('inf')  # Повертаємо нескінченний порядок для неквадратних матриць
    
    identity = matrix.parent().identity_matrix()  
    current = matrix
    for n in range(1, max_iterations + 1):
        if current == identity:
            return n
        current *= matrix  
    return float('inf')  # Якщо порядок перевищує max_iterations, вважаємо його нескінченним

for g in unique_elements:
    element_matrix = Matrix(g)
    order = compute_order(element_matrix)
    print(f"Element {element_matrix} has order: {order}")




```
